```python
import pandas as pd

# Corrected URL for direct download from Google Drive
url = "https://drive.google.com/uc?export=download&id=1IDu3QPY8fFHNTmedft5chiY8O3ltnAhz"
df = pd.read_csv(url)

SIZE = 50
hash_table = [None] * SIZE

def hash_code(key):
    return key % SIZE

def insert(key, value):
    index = hash_code(key)
    while hash_table[index] is not None:
        index = (index + 1) % SIZE
    hash_table[index] = (key, value)   # store tuple

def search(key):
    index = hash_code(key)
    while hash_table[index] is not None:
        if hash_table[index][0] == key:
            return hash_table[index]
        index = (index + 1) % SIZE
    return None

# Corrected iteration over DataFrame rows to populate the hash table
for _, row in df.iterrows():
    insert(int(row["StudentID"]), row["Name"])

print(search(1001))
```
